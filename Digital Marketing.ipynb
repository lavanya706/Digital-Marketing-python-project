```python
import pandas as pd
```

```python
data=pd.read_csv("_Inputdata.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
columns_to_keep = ['Campaign name', 'Day', 'Amount spent (INR)', 'CTR (link click-through rate)', 'Link clicks', 'Landing page views', 'Results']
data = data[columns_to_keep]
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['CTR (link click-through rate)'].fillna(data['CTR (link click-through rate)'].median(), inplace=True)
data['Link clicks'].fillna(data['Link clicks'].median(), inplace=True)
data['Landing page views'].fillna(data['Landing page views'].median(), inplace=True)
data['Results'].fillna(data['Results'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
def calculate_campaign_metrics(data):
    ad_spent = data['Amount spent (INR)']
    no_of_leads = data['Results']
    cost_per_lead = ad_spent / no_of_leads
    ctr = data['CTR (link click-through rate)']
    link_click = data['Link clicks']
    lp_views = data['Landing page views']
    member_passed = (lp_views / link_click) * 100
    lp_conversion = (no_of_leads / lp_views) * 100
    
    
    campaign_metrics = pd.DataFrame({
        "AD Spend": ad_spent,
        "No of Leads": no_of_leads,
        "Cost Per Lead": cost_per_lead,
        "CTR": ctr,
        "Link Click": link_click,
        "LP Views": lp_views,
        "Member Passed": member_passed,
        "LP Conversion %": lp_conversion
    })

    return campaign_metrics
```

```python
calculate_campaign_metrics(data)
```

```python

result_data = pd.DataFrame(columns=["Date", "Campaign Name", "AD Spend", "No of Leads", "Cost Per Lead",
                                    "CTR", "Link Click", "LP Views", "Member Passed", "LP Conversion %"])

for campaign_name in data['Campaign name'].unique():
    campaign_data = data[data['Campaign name'] == campaign_name]
    
    campaign_dates = campaign_data['Day'].unique()

    for date in campaign_dates:
        date_data = campaign_data[campaign_data['Day'] == date]
    
        campaign_metrics = calculate_campaign_metrics(date_data)
        
        if campaign_metrics is not None:
            campaign_metrics['Date'] = date
            campaign_metrics['Campaign Name'] = campaign_name
            result_data = pd.concat([result_data, campaign_metrics], ignore_index=True)
result_data.to_csv("campaign_result.csv", index=False, columns=['Date', 'Campaign Name', 'AD Spend', 'No of Leads', 'Cost Per Lead',
                                                                           'CTR', 'Link Click', 'LP Views', 'Member Passed', 'LP Conversion %'])
```
